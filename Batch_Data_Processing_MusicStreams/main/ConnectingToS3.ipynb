```python
import os
import pandas as pd
import boto3
from dotenv import load_dotenv

load_dotenv()
os.environ["AWS_ACCESS_KEY_ID"] = os.environ.get('AWS_ACCESS_KEY_ID')
os.environ["AWS_SECRET_ACCESS_KEY"] = os.environ.get('AWS_SECRET_ACCESS_KEY')
os.environ["AWS_DEFAULT_REGION"] = os.environ.get('AWS_DEFAULT_REGION')
```

##### 1. Connect to S3 boto3 client

```python
s3_client = boto3.client('s3')
```

##### 2. List the object in respective buckets

```python
objects = s3_client.list_objects_v2(Bucket = 'de-airflow-sansdags', Prefix = "streams/")
```

##### 3. Reading the Objects in S3 location

```python
response = s3_client.get_object(Bucket = "de-airflow-sansdags", Key = 'streams/streams1.csv')
data = pd.read_csv(response['Body'])
data
```
